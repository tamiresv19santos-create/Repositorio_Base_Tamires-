### üè¢ Desafio: Sistema de Cadastro de Funcion√°rios com Listas e Loop for

üìå Contexto do Desafio

Voc√™ foi contratado(a) pela **MSCorp** para desenvolver um sistema de cadastro de funcion√°rios usando listas. O programa deve:

‚úÖ Cadastrar alguns funcion√°rios (quantos o usu√°rio quiser)

‚úÖ Armazenar nome, cargo e sal√°rio de cada um em listas separadas.

‚úÖ No final, exibir um relat√≥rio organizado com todos os dados.

### üìã Requisitos do Sistema


1Ô∏è‚É£ Estrutura dos Dados
Usaremos 3 listas separadas para armazenar os dados:



### Meu c√≥digo

```python
nomes = []
cargos = []
salarios = []

quantidade = int(input("Quantos funcion√°rios deseja castrar??"))
for i in range(quantidade):
    print(f"\n üìã Cadastro do funcionario ")
    nome = input("Nome:")
    cargo = input("Cargo:")
    salario = input("Salario:")

    nomes.append(nome)
    cargos.append(cargo)
    salarios.append(salario)
    print("\n --- RELAT√ìRIO DE FUNCION√ÅRIOS ---")



for i in range(quantidade):
         print(f"Nome:{nomes[i]} | Cargo:{cargos[i]} | Sal√°rio:R$ {salarios[i]}: ")
print(f"A quantidade de funcionarios cadastrados {quantidade}")
```

```python
# lista para armazenar os dados 
nomes = []
cargos = []
salarios = []

quantidade = int(input("Quantos funcionarios deseja cadastrar??"))

for i  in range(quantidade):
    print(f"\nüìã cadastro dp {i+1}¬∫ funcionario:")
    nome = input(" Nome Completo:").title()
    cargo = input("cargo: ").capitalize()
    salario = float(input("salario R$"))

    # adiciona os dados √†s listas
    nomes.append(nome)
    cargos.append(cargo)
    salarios.append(salario)

    # exibe o relatorio final
    print("\n üåü**Relat√≥rio de Funcion√°rios**üåü")
    print("-" * 60)
    print(f"{'Nome':<20} | {'Cargo':<20} | R$ {'salario':>10}")
    print("-" * 60)
for i in range(len(nomes)):
      print(f"{nomes[i]:<20} | {cargos[i]:<20} | R$ {salarios[i]:>8.2f}")
print("-" * 60)
print(f"Total de funcionarios contratados: {len(nomes)}")
```

```python
### Importando a biblioteca 
import matplotlib.pyplot as plt
# Gr√°fico 1: Distribui√ß√£o de Sal√°rios
# Cria um subplot (1 linha, 2 colunas, posi√ß√£o 1)
plt.subplot(1, 2, 1)
# Cria um gr√°fico de barras com os nomes no eixo X e sal√°rios no eixo Y
plt.bar(nomes, salarios, color='skyblue')  # Barras azuis claras
# Configura√ß√µes do gr√°fico
plt.title('Sal√°rios dos Funcion√°rios')  # T√≠tulo do gr√°fico
plt.xlabel('Funcion√°rios')             # R√≥tulo do eixo X
plt.ylabel('Sal√°rio (R$)')             # R√≥tulo do eixo Y
plt.xticks(rotation=45)               # Rotaciona os r√≥tulos do eixo X em 45 graus
plt.grid(axis='y')                    # Mostra linhas de grade apenas no eixo Y
# Gr√°fico 2: Distribui√ß√£o por Cargo (contagem)
# Cria o segundo subplot (posi√ß√£o 2)
plt.subplot(1, 2, 2)
# Prepara dados para o gr√°fico de pizza
cargos_unicos = list(set(cargos))  # Remove duplicatas usando set() e converte para lista
contagem = [cargos.count(cargo) for cargo in cargos_unicos]  # Conta quantos de cada cargo
# Cria gr√°fico de pizza
plt.pie(contagem, 
        labels=cargos_unicos,       # R√≥tulos para cada fatia
        autopct='%1.1f%%',          # Mostra porcentagens com 1 casa decimal
        colors=['lightcoral', 'lightgreen', 'gold'])  # Cores para cada fatia
plt.title('Distribui√ß√£o por Cargo')  # T√≠tulo do gr√°fico
# Ajusta o layout para evitar sobreposi√ß√£o
plt.tight_layout()
# Mostra os dois gr√°ficos (de barras e pizza) juntos
plt.show()
# Gr√°fico 3: Histograma de Faixas Salariais (extra)
# Cria uma nova figura com tamanho espec√≠fico (8x5 polegadas)
plt.figure(figsize=(8, 5))
# Define os intervalos (bins) para o histograma
faixas = [0, 3000, 5000, 7000, 9000, 11000]
# Cria o histograma
plt.hist(salarios, 
         bins=faixas,               # Define os intervalos salariais
         edgecolor='black',         # Cor da borda das barras
         color='lightseagreen')     # Cor de preenchimento das barras
# Configura√ß√µes do gr√°fico
plt.title('Distribui√ß√£o por Faixa Salarial')  # T√≠tulo
plt.xlabel('Faixas Salariais (R$)')          # R√≥tulo eixo X
plt.ylabel('Quantidade de Funcion√°rios')     # R√≥tulo eixo Y
plt.xticks(faixas)                          # Define os ticks do eixo X
# Mostra o histograma
plt.show()
```
